## Analytical RS+FD

Parameters

```python
# Range of epsilon used in the experiments
lst_eps = [round(np.log(val),4) for val in range(2,8)] 

# Parameters of Adult Dataset. 
df = pd.read_csv('db_ACSEmployement.csv')
n = df.shape[0] #number of users
d = df.shape[1] #number of attributes
beta = 1/d #sampling rate
K = [df[att].unique().shape[0] for att in df.columns] #number of values per attribute
```

```python
# from https://github.com/hharcolezi/ldp-protocols-mobility-cdrs/tree/main/papers/%5B2%5D
def Var_RSpFD_GRR(n, d, k, p, q): 
    f_vi = 0 # approximate variance
    
    sig_grr = (1/d) * (q + f_vi*(p-q) + (d-1)/k)
    
    var_grr = ((d**2 * sig_grr * (1-sig_grr))/(n*(p-q)**2))
    
    return var_grr

def Var_RSpFD_UE_r(n, d, k, p, q): 
    f_vi = 0 # approximate variance
    
    sig_ue_r = (1/d) * (f_vi * (p-q) + q + (d-1) * ((p/k) + ((k-1)/(k))*q) )
    
    var_ue_r = ((d**2 * sig_ue_r * (1-sig_ue_r))/(n*(p-q)**2))
    
    return var_ue_r
```

```python
# Initialize list of averaged variance values per epsilon
lst_var_RSpFD_GRR = []
lst_var_RSpFD_SUE_r = []
lst_var_RSpFD_OUE_r = []

for eps in lst_eps:
    eps_l = np.log((np.exp(eps) - 1 + beta)/beta) #amplified epsilon
    
    # Initialize list of variance values per attribute
    lst_var_att_grr = []
    lst_var_att_sue_r = []
    lst_var_att_oue_r = []

    for k in K: # for each attribute
        
        p = np.exp(eps_l) / (np.exp(eps_l) + k - 1) # GRR
        q = (1 - p) / (k - 1) # GRR
        
        p_sue = np.exp(eps_l/2) / (np.exp(eps_l/2) + 1) #SUE
        q_sue = 1 - p_sue  #SUE
        
        p_oue = 1/2 #OUE
        q_oue = 1 / (np.exp(eps_l)+1) #OUE
        
        lst_var_att_grr.append(Var_RSpFD_GRR(n, d, k, p, q))
           
        lst_var_att_sue_r.append(Var_RSpFD_UE_r(n, d, k, p_sue, q_sue))
        
        lst_var_att_oue_r.append(Var_RSpFD_UE_r(n, d, k, p_oue, q_oue))           
           
    # We'll average the MSE over the number of attributes
    lst_var_RSpFD_GRR.append(np.mean(lst_var_att_grr))
    lst_var_RSpFD_SUE_r.append(np.mean(lst_var_att_sue_r))
    lst_var_RSpFD_OUE_r.append(np.mean(lst_var_att_oue_r))
```

## Analytical RS+RFD

```python
def Var_RSpRFD_GRR(n, d, k, p, q, f_prior): 
    f_vi = 0 # approximate variance
    
    sig_grr = (1/d) * (q + f_vi*(p-q) + (d-1)*f_prior)
    
    var_grr = ((d**2 * sig_grr * (1-sig_grr))/(n*(p-q)**2))
    
    return np.mean(var_grr)


def Var_RSpRFD_UE_r(n, d, k, p, q, f_prior): 
    f_vi = 0 # approximate variance
    
    sig_ue_r = (1/d) * (f_vi * (p-q) + q + (d-1)*(f_prior*(p-q)+q))
    
    var_ue_r = ((d**2 * sig_ue_r * (1-sig_ue_r))/(n*(p-q)**2))
    
    return np.mean(var_ue_r)
```

Incorrect priors DIR

```python
nb_seed = 100

prior = 'Incorrect_DIR'

seed_RSpRFD_GRR = []
seed_RSpRFD_SUE_r  = []
seed_RSpRFD_OUE_r = []

for seed in range(nb_seed):

    freq_prior = []
    for k in K:

        freq = np.random.dirichlet(np.ones(k), size=1)[0]
        freq = freq/sum(freq)
        freq_prior.append(freq)
    
    # Initialize list of averaged variance values per epsilon
    lst_var_RSpRFD_GRR = []
    lst_var_RSpRFD_SUE_r = []
    lst_var_RSpRFD_OUE_r = []
    
    for eps in lst_eps:
        eps_l = np.log((np.exp(eps) - 1 + beta)/beta) #amplified epsilon
        # Initialize list of variance values per attribute
        lst_var_att_grr = []
        lst_var_att_sue_r = []
        lst_var_att_oue_r = []

        idx_k = 0
        for k in K: # for each attribute

            p = np.exp(eps_l) / (np.exp(eps_l) + k - 1) # GRR
            q = (1 - p) / (k - 1) # GRR

            p_sue = np.exp(eps_l/2) / (np.exp(eps_l/2) + 1) #SUE
            q_sue = 1 - p_sue  #SUE

            p_oue = 1/2 #OUE
            q_oue = 1 / (np.exp(eps_l)+1) #OUE

            lst_var_att_grr.append(Var_RSpRFD_GRR(n, d, k, p, q, freq_prior[idx_k]))

            lst_var_att_sue_r.append(Var_RSpRFD_UE_r(n, d, k, p_sue, q_sue, freq_prior[idx_k]))

            lst_var_att_oue_r.append(Var_RSpRFD_UE_r(n, d, k, p_oue, q_oue, freq_prior[idx_k]))           
            idx_k+=1

        # We'll average the MSE over the number of attributes
        lst_var_RSpRFD_GRR.append(np.mean(lst_var_att_grr))
        lst_var_RSpRFD_SUE_r.append(np.mean(lst_var_att_sue_r))
        lst_var_RSpRFD_OUE_r.append(np.mean(lst_var_att_oue_r))
        
    seed_RSpRFD_GRR.append(lst_var_RSpRFD_GRR)
    seed_RSpRFD_SUE_r.append(lst_var_RSpRFD_SUE_r)
    seed_RSpRFD_OUE_r.append(lst_var_RSpRFD_OUE_r)
```
